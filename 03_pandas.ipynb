# Pandas

Es una librería open source para el análisis y preocesamiento de datos, que además facilita la creación de estructuras de datos de una forma muy sencilla.

Aprenderemos a leer archivos y transformarlos en estructuras de datos que nos van a facilitar mucho la vida a la hora de analizarlos.

## Archivos CSV (Local y Web)

```python
import pandas as pd
```

```python
raw_data = {'first_name':['Jason','Molly','Tina','Jake','Amy'],
           'last_name':['Miller','Jacobson','.','Milner','Cooze'],
           'age':[42,52,36,24,73],
           'preTestScore':[4,24,31,".","."],
           'postTestScore':["25,000","94,000",57,62,70]}
df = pd.DataFrame(raw_data)
df
```

## Guardar nuestro datos como csv

```python
df.to_csv('C://Users//Samue//x-anaconda//datos//EjemploPandas.csv')
```
